```python
# Import the data from the three files
import pandas as pd
url = '/Users/skyler/Documents/GitHub/lab-dw-data-structuring-and-combining/file1.csv'
url1 = '/Users/skyler/Documents/GitHub/lab-dw-data-structuring-and-combining/file2.csv'
url2 = '/Users/skyler/Documents/GitHub/lab-dw-data-structuring-and-combining/file3.csv'
insurance_data = pd.read_csv(url)
new_data1 = pd.read_csv(url1)
new_data2 = pd.read_csv(url2)

insurance_data
```

```python
new_data1.columns = insurance_data.columns.str.lower().str.replace(' ', '_')
new_data2.columns = insurance_data.columns.str.lower().str.replace(' ', '_')
insurance_data.columns = insurance_data.columns.str.lower().str.replace(' ', '_')
```

```python
combined_data = pd.concat([insurance_data, new_data1, new_data2], axis=0)
```

```python

combined_data = combined_data.rename(columns={'st': 'state'})
combined_data["gender"] = combined_data["gender"].replace({"Femal": "F", "femal": "F","female": "F", "Male": "M"})
combined_data['state'] = combined_data['state'].replace({"AZ":"Arizona", "WA": "Washington "})
combined_data['education'] = combined_data['education'].replace({"Bachelors":"Bachelor"})
combined_data['customer_lifetime_value'] = combined_data['customer_lifetime_value'].str.replace("%", '')
```

```python
combined_data
```

```python
combined_data['customer_lifetime_value'] = pd.to_numeric(combined_data['customer_lifetime_value'], errors='coerce')
combined_data.dtypes
```

```python
def extract_middle_value(complaints):
    if isinstance(complaints, str):  # Ensure the value is a string
        parts = complaints.split('/')
        if len(parts) == 3:
            return parts[1]
    return complaints 

combined_data['number_of_open_complaints'] = combined_data['number_of_open_complaints'].apply(extract_middle_value)
combined_data
```

```python

combined_data = combined_data.dropna(how='all')
combined_data.isna().sum()
```

```python
combined_data['customer_lifetime_value'] = combined_data['customer_lifetime_value'].fillna(combined_data['customer_lifetime_value'].mean())

combined_data['gender'].fillna(combined_data['gender'].mode()[0], inplace=True)
combined_data.isna().sum()
```

```python
numeric_columns= combined_data.select_dtypes(include=['float64']).columns
combined_data[numeric_columns] = combined_data[numeric_columns].astype(int)
combined_data.dtypes
```

```python
combined_data.duplicated().sum()
combined_data.reset_index(drop=True, inplace=True)
combined_data
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
